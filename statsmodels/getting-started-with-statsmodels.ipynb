## Getting Started with Statsmodels

```python
import statsmodels.api as sm
import pandas as pd
pd.set_option('display.max_columns', None)
from patsy import dmatrices
```

```python
df = sm.datasets.get_rdataset("Guerry", "HistData").data
```

```python
df
```

```python
variables = ["Department", "Lottery", "Literacy", "Wealth", "Region"]
df = df[variables]
df
```

```python
df.info()
```

```python
df = df.dropna()
df.info()
```

```python
y, X = dmatrices('Lottery ~ Literacy + Wealth + Region', 
                 data = df, 
                 return_type='dataframe')
```

```python
y
```

```python
type(y)
```

```python
X.head()
```

### Model fit and summary

```python
# Describe the model
mod = sm.OLS(y, X)

# Fit model
res = mod.fit()

# Summarize model
res.summary()
```

```python
print(res.params)
print("==================")
print(res.rsquared)
```

#### Diagnostics and specification test

```python
sm.stats.linear_rainbow(res)
```

```python
print(sm.stats.linear_rainbow.__doc__)
```

```python
sm.graphics.plot_partregress('Lottery', 'Wealth', ['Region', 'Literacy'], data=df, obs_labels = False);
```
