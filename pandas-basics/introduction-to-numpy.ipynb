## NumPy

* NumPy (or Numpy) is a Linear Algebra Library for Python, the reason it is so important for Data Science with Python is that almost all of the libraries in PyData Ecosystem rely on NumPy as one of their main building blocks.

* Numpy is also incredibly fast, as it has bindings to C libraries.

* We will only learn the basics of NumPy

### Using NumPy

Import it as a library:

```python
import numpy as np
```

Numpy has many built-in functions and capabilities. We won't cover them all but instead we will focus on some of the most important aspects of Numpy: vectors, arrays, matrices, and number generation. Let's start by discussing arrays.

### Numpy arrays

NumPy arrays are the main way we will use Numpy throughout the course. Numpy arrays essentially come in two flavors: vectors and matrices. Vectors are strictly 1-d arrays and matrices are 2-d (but you should note a matrix can still have one row or one column).

Let's begin our introduction by exploring how to create NumPy arrays.

#### NumPy Indexing and Selection

In this lecture we will discuss how to select elements or groups of elements from an array.

```python
# Creating sample array
arr = np.arange(0,11)
```

```python
# Show
arr
```

#### Bracket Indexing and Selection

The simplest way to pick one or some elements of an array looks very similar to python lists:

```python
# Get a value at an index
arr[8]
```

```python
# Get values in a range
arr[1:5]  # slice
```

```python
# Get values in a range
arr[0:5]
```

#### Broadcasting

Numpy arrays differ from a normal Python list because of their ability to broadcast:

```python
# Setting a value with index range (Broadcasting)

arr[0:5]=100

# Show 
arr
```

```python
# Reset array
arr = np.arange(0,11)

# Show
arr
```

#### Broadcasting

```python
# Important notes on slices
slice_of_arr = arr[0:6]

# Show slice
slice_of_arr
```

```python
# Change Slice
slice_of_arr[:] = 99

# Show Slice again
slice_of_arr
```

#### Broadcasting (DANGERS)

Now note the changes also occur in our original array!

```python
arr
```

Data is not copied, it's a view of the original array! This avoids memory problems!

#### Copying

```python
# To get a copy, need to be explicit
arr_copy = arr.copy()

arr_copy
```

```python
slice_of_arr_2 = arr_copy[0:5]
slice_of_arr_2
```

```python
slice_of_arr_2[:] = 1000

slice_of_arr_2
```

```python
arr
```

#### Indexing a 2D array (matrices)

The general format is **arr_2d[row][col]** or **arr_2d[row][col]**. I recommend usually the common notation for clarity.

```python
arr_2d = np.array(([5,10,15], [20,25,30], [35,40,45]))

# Show 
arr_2d
```

```python
# Indexing row
arr_2d[1]
```

#### Indexing a 2D array (matrices)

```python
# Format is arr_2d[row][col] or arr_2d[row,col]

# Getting individual element value
arr_2d[1][0]
```

```python
# Getting individual element value
arr_2d[1,0]
```

#### Indexing a 2D array (matrices)

```python
# 2d array slicing

# Shape (2,2) from top right corner
arr_2d[:2,1:]
```

```python
# Shape bottom row
arr_2d[2]
```

```python
# Shape bottom row
arr_2d[2,:]
```

#### Fancy indexing

Fancy indexing allows you to select entire rows or columns out of order, to show this, let's quickly build out a numpy array:

```python
# Set up matrix
arr2d = np.zeros((10,10))
```

```python
# Length of array
arr_length = arr2d.shape[1]
```

```python
# Set up array
for i in range(arr_length):
    arr2d[i] = i

arr2d
```

```python
arr2d[[2,4,6,8]]
```

```python
# Allows in any order
arr2d[[6,4,2,7]]
```

```python
# Calculating exponential (e^)
np.exp(arr)
```
