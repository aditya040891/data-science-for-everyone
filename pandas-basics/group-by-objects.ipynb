## Grouping

By "group by" we are referring to a process involving one or more of the following steps:

* **Splitting** the data into groups based on some criteria
* **Applying** a function to each group independently
* **Combining** the results into a data structure

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'A': ['foo', 'bar', 'foo', 'bar', 'foo', 'bar', 'foo', 'foo'],
    'B': ['one', 'one', 'two', 'three', 'two', 'two', 'one', 'three'],
    'C': np.random.randn(8),
    'D': np.random.randn(8)
})
```

```python
df
```

Grouping and then applying the `sum()` function to the resulting groups.

```python
df.groupby('A')
```

```python
df.groupby('A').sum(numeric_only=True)
```

Grouping by multiple columns forms a hierarchical index, and again we can apply the sum function.

```python
df.groupby(['A', 'B']).sum()
```

- Aggregation: compute a summary statistic (or statistics) for each group. Some examples:
* Compute group sums or means.<br>
* Compute group sizes / counts.<br>
<br>
- Transformation: perform some group-specific computations and return a like-indexed object. Some examples:<br>
* Standardize data (zscore) within a group.<br>
* Filling NAs within groups with a value derived from each group.<br>
<br>
<br>
- Filtration: discard some groups, according to a group-wise computation that evaluates True or False. Some examples:<br>
* Discard data that belongs to a groups with only a few members.<br>
* Filter out data based on the group sum or mean.<br>

### Example

```python
df = pd.DataFrame(
    [
        ('bird', 'Falconiformes', 389.0),
        ('bird', 'Psittaciformes', 24.0),
        ('mammal', 'Carnivora', 80.2),
        ('mammal', 'Primates', np.nan),
        ('mammal', 'Carnivora', 58),
    ],
    index = ['falcon', 'parrot', 'lion', 'monkey', 'leopard'],
    columns = ('class', 'order', 'max_speed')
)
```

```python
df
```

```python
g_class = df.groupby('class') # axis = 0
g_order = df.groupby('order', axis='columns') # axis = 1, axis="columns"
g_both = df.groupby(['class', 'order'])
```

```python
g_class.mean(numeric_only=True)
```

```python
g_both.mean()
```

```python
data = {
    'school': ['HSU', 'HSU', 'HSU', 'OBU', 'OBU', 'SIU', 'SIU', 'SEU', 'SEU'],
    'professor': ['Bob', 'Jeff', 'Angela', 'Susan', 'Albert', 'Zelda', 'Alexa', 'Heather', 'Rebecca'],
    'publication': [10, 2, 30, 25, 0, 80, 4, 30, 15],
    'age': [76, 26, 30, 25, 50, 80, 40, 30, 50]
}
df = pd.DataFrame(data)
```

```python
df
```

### groupby()

* use the .groupby() method to group rows together based off a column name
* Create a group based off school

```python
df.groupby('school') # creates a DataFrameGroupBy object
```

```python
by_school = df.groupby('school')
```

#### Groupby and Aggregation methods

* Use the aggregation methods on the grouped object

```python
by_school.mean(numeric_only=True)
```

```python
by_school.std(numeric_only=True) # standard deviation
```

```python
by_school.min()
```

```python
by_school.max()
```

```python
by_school.count()
```

#### Groupby+describe method

```python
by_school.describe()
```
