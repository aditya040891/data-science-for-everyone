## Reshaping and Pivoting DataFrames with Pandas

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'date': ['2000-01-03', '2000-01-04', '2000-01-05', '2000-01-03', '2000-01-04', 
    '2000-01-05', '2000-01-03', '2000-01-04', '2000-01-05', '2000-01-03', '2000-01-04', '2000-01-05'],
    'variable': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C', 'D', 'D', 'D'],
    'value': [0.794309, 0.427748, -1.544155, 0.128190, -0.351577, 0.839166, -0.563507, 0.744192, 0.016040, 0.287683, 0.224715, -0.583553]
})
```

```python
df
```

```python
df[df['variable']=='B']
```

```python
df.pivot(index='date', columns='variable', values='value')
```

```python
df['value_2'] = df['value']*2
```

```python
df
```

```python
df_pivot_vals = df.pivot(index='date', columns='variable')
```

```python
df_pivot_vals
```

```python
df_pivot_vals['value_2']
```

#### Reshaping by stacking and unstacking

* **stack**: "pivot" a level of the column labels, returning a DataFrame with an index with a new inner most level of row labels.
* **unstack**: (inverse operation of stack) "pivot" a level of the row index to the column axis, producing a reshaped DataFrame with a new innet most level of column labels.

```python
df
```

```python
df2 = pd.DataFrame(
    data = {
        'A': [-0.594045, 0.812831, -1.492472, 0.422326],
        'B': [0.989808, 0.744916, -0.668162, -0.545292]    
    }
)
```

```python
df2
```

```python
df2.index = pd.MultiIndex.from_product([['bar', 'baz'], ['one', 'two']], names=['first', 'second'])
```

```python
df2
```

```python
stacked = df2.stack()
```

```python
stacked
```

```python
stacked.unstack(2)
```

```python
stacked.unstack(1)
```

```python
stacked.unstack(0)
```

```python
stacked.unstack('second')
```

```python
df3 = pd.DataFrame([[0.160287, 1.015534, 1.030470, 0.508472],
                  [1.368757, -1.447397, -0.246379, 0.153493],
                  [1.230501, -2.224046, 1.341282, 0.186118],
                  [-0.082465, -1.045151, 0.432698, -0.063779]])
```

```python
df3
```

```python
df3.columns = pd.MultiIndex.from_arrays([
    ['A', 'B', 'A', 'B'],
    ['cat', 'cat', 'dog', 'dog'],
    ['long', 'long', 'short', 'short']],
    names=['exp', 'animal', 'hair_length']
)
```

```python
df3
```

```python
df3.stack(level=['animal', 'hair_length']) # string index / named index
```

```python
df3.stack(level=[1,2]) # numeric index
```

#### Melt

```python
cheese = pd.DataFrame(
    {
        'first': ['John', 'Mary'],
        'last': ['Doe', 'Bo'],
        'height': [5.5, 6.0],
        'weight': [130, 150]
    }
)
```

```python
cheese
```

```python
cheese.melt(id_vars=['first', 'last'])
```

```python
cheese.melt(id_vars=['first', 'last'], var_name='quant', value_name='metrics')
```
