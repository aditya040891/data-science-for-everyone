## Financial Data Science

### 2D Static Plotting

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42) # Fixes the seed for reproducibility
```

```python
y = np.random.standard_normal(100)  # Draw 100 rand numbers
x = np.arange(len(y)) # Sets up int
```

```python
y
```

```python
x
```

```python
plt.plot(x, y)
```

```python
plt.plot(y)
```

```python
plt.plot(y.cumsum())
plt.grid(True)
plt.axis('equal');
```

```python
plt.plot(y.cumsum())
plt.xlim(-1, 20)
plt.ylim(np.min(y.cumsum()) -1,
        np.max(y.cumsum()) + 1)
plt.grid(True);
```

```python
plt.figure(figsize=(10, 5))
plt.plot(y.cumsum(), 'g', lw=1.5)
plt.plot(y.cumsum(), 'r1')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Very Simple Plot');
```

### 2D plotting

```python
y = np.random.standard_normal((100, 2)).cumsum(axis=0)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(y, lw=1.5)
plt.plot(y, 'ro')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Simple 2D');
```

```python
plt.figure(figsize=(10, 5))
plt.plot(y[:,0], 'g', label='First', lw=1.5)
plt.plot(y[:,1], 'b', label='Second', lw=1.5)
plt.legend()
plt.plot(y, 'ro')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Simple 2D');
```

```python
y[:,0] = y[:,0]*100
```

```python
fig, ax1 = plt.subplots()
# plt.figure(figsize=(10, 5))
plt.plot(y[:,0], 'g', label='First', lw=1.5)
plt.legend(loc=8)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Simple 2D')

ax2 = ax1.twinx()
plt.plot(y[:,1], 'b', label='Second', lw=1.5)
plt.legend(loc=0)
plt.ylabel('Value 2');
```

```python
plt.figure(figsize=(10,5))
plt.subplot(211)
plt.plot(y[:, 0], 'g', label='First')
plt.legend(loc=0)
plt.ylabel('Value')
plt.title('Simple 2D Plots')

plt.subplot(212)
plt.plot(y[:,1], label='Second')
plt.legend(loc=0)
plt.xlabel('Index')
plt.ylabel('Value');
```

```python
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.plot(y[:,0], label='Stock')
plt.legend(loc=0)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Stock Value')

plt.subplot(122)
plt.bar(np.arange(len(y)), y[:,1], width=0.5, color="r", label="Volume")
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('Vol. of Stock');
```
