### Advanced Financial Data Gathering

```python
# Major indices
import requests
import time

url = "https://finance.yahoo.com/markets/world-indices"

headers = {
    "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chromium/145.0.7632.109 Safari/537.36"
}

time.sleep(2)   # wait 2 seconds before sending request

response = requests.get(url, headers=headers)

major_indices = pd.read_html(response.text)[0]
```

```python
major_indices
```

```python
ticker_list = major_indices['Symbol'].str.replace('^', '').str.lower().to_list()
ticker_list
```

```python
ticker_list = ticker_list[:-1]
ticker_list
```

```python
ticker_data = yf.Tickers(ticker_list)
```

```python
ticker_data
```

```python
df = yf.download(ticker_list, period='1d', start='2022-01-13', end='2024-03-10')
```
