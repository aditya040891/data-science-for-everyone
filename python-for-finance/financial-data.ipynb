## Financial Data 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import yfinance as yf
pd.set_option('display.max_columns', None)
```

```python
stocks = dict(
    AAPL = "Apple Stock",
    AMZN = "Amazon Stock",
    CNY = "CNY/USD Stock",
    GDX = "Gold Price",
    GLD = "SPDR Gold Trust",
    GS = "Goldman Sachs Stock",
    INTC = "Intel Stock",
    MSFT = "Microsoft Stock",
    SPY = "S&P500"
)
```

```python
stocks
```

```python
list(stocks.keys())
```

```python
df = yf.download(list(stocks.keys()), period='max')
```

```python
df.head()
```

```python
df.info()
```

```python
close = df['Close']
```

```python
close.head()
```

```python
close.dropna().plot(subplots=True, figsize=(12, 10));
```

```python
close = close.dropna()
```

```python
close
```

```python
for key, value in stocks.items():
    print(f"{key:8s}|{value}")
```

```python
close.agg(['min', 'mean', 'std', 'median', 'max']).round(2)
```

##### Change over time

```python
close.diff().mean().head()
```

```python
close.pct_change().round(3).head()
```

```python
close.pct_change().mean().plot(kind='bar', figsize=(5,5));
```

```python
print(close.head())
print("----------------")
print(close.shift(1).head())
```

```python
rets = np.log(close/close.shift(1))
```

```python
rets.cumsum().apply(np.exp).plot(figsize=(10,5));
```

### Resampling

```python
close.resample('1W', label='right').last().head()
```

```python
close.resample('1ME', label='right').last().head()
```

```python
rets.cumsum().apply(np.exp).resample('1ME', label='right').last().plot(figsize=(10,5));
```
