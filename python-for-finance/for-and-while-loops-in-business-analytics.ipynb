### Quickmart sales analysis

```python
# Monthly sales data per store (in thousands of dollars)
sales_data = {
    "Downtown": [120.5, 115.8, 131.2, 140.5, 150.2, 160.1, 155.3, 148.9, 152.5, 160.8, 165.2, 180.3],
    "Suburb Mall": [85.6, 90.2, 93.5, 100.8, 110.5, 115.7, 120.2, 118.5, 125.6, 130.2, 140.8, 155.5],
    "Westside": [95.2, 88.7, 92.3, 100.5, 105.8, 110.2, 115.7, 120.5, 125.8, 130.2, 135.5, 145.8],
    "University": [55.3, 60.2, 65.8, 70.5, 65.2, 50.1, 45.2, 55.8, 80.5, 85.9, 90.2, 95.3]
}
```

```python
locations = list(sales_data.keys())

months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
```

```python
# Monthly totals (nested for loop)
monthly_totals = [] # 12 floats

for m in range(12):  # outer loop over months
    month_sum = 0.0
    for store in sales_data: # inner loop over stores
        month_sum += sales_data[store][m]
    monthly_totals.append(round(month_sum, 2))

# Best / Worst months range(len()) + conditionals
highest_month_index = 0
lowest_month_index = 0
highest_sales = 0
lowest_sales = float('inf')
```

```python
for i in range(1, len(monthly_totals)):
    if monthly_totals[i] > highest_sales:
        highest_sales = monthly_totals[i]
        highest_month_index = i
    if monthly_totals[i] < lowest_sales:
        lowest_sales = monthly_totals[i]
        lowest_month_index = i
```

```python
# Average monthly sales (loop accumulator)
sum_totals = 0.0

for val in monthly_totals:
    sum_totals += val

average_monthly_sales = round(sum_totals/12, 2)
```

```python
# Best performing store (loop over stores)
# best_store, best_store_sales (float)
best_stores = ""
best_store_sales = 0.0

for store in sales_data:
    total_sales = sum(sales_data[store])
    if total_sales > best_store_sales:
        best_store_sales = round(total_sales, 2)
        best_store = store
```

```python
# Monthly performance report
# While loop + above/below average
# performance_report (list of strings)
performance_report = []
i = 0
while i < 12:
    total = monthly_totals[i]
    indicator = "Above Average" if total >= average_monthly_sales else "Below Average"
    line = f"{months[i]}: ${total:.2f} - {indicator}"
    performance_report.append(line)
    i += 1
```

```python
# Longest consecutive growth streak
# longest_growth_streak (int), growth_streak_start (int)
# monthly_total[i] > monthly_totals[i-1]

longest_growth_streak = 1  # number of months in the longest increasing run
growth_streak_start = 0  # starting month index of that run

current_streak_len = 1
current_streak_start = 0

for r in range(1,12):
    if monthly_totals[r] > monthly_totals[r-1]:
        current_streak_len += 1
    else:
        # streak broken; check if the current one is the best
        if current_streak_len > longest_growth_streak:
            longest_growth_streak = current_streak_len
            growth_streak_start = current_streak_start

        # Reset for a new streak starting at i
        current_streak_len = 1
        current_streak_start = r

# Finally check in case best streak ends in December
if current_streak_len > longest_growth_streak:
    longest_growth_streak = current_streak_len
    growth_streak_start = current_streak_start
```

```python
print("QUICKSMART SALES ANALYSIS SUMMARY")
print("="*62)
print(f"Best Month: {months[highest_month_index]} - ${highest_sales:.2f}")
print(f"Worst Month: {months[lowest_month_index]} - ${lowest_sales:.2f}")
print(f"Average Monthly Sales: ${average_monthly_sales:.2f}")
print(f"Best Peforming Store: {best_store} - ${best_store_sales:.2f}")
print(f"Longest Growth Streak: {longest_growth_streak} months starting in {months[growth_streak_start]}")

print("\nMonthly Performance Report")
print("-" * 62)

for line in performance_report:
    print(line)
```
