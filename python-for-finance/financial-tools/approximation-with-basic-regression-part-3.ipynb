## Financial Tools: Approximation with Basic Regression Part 3

### Noisy Data

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x) + 0.5 * x
```

```python
def create_plot(x, y, styles, labels, axlabels):
    """Simplification function for plotting"""
    plt.figure(figsize=(10,6))
    for i in range(len(x)):
        plt.plot(x[i], y[i], styles[i], label=labels[i])
        plt.xlabel(axlabels[0])
        plt.ylabel(axlabels[1])
    plt.legend(loc=0)
```

```python
xn = np.linspace(-2*np.pi, 2*np.pi, 50) # x values
xn = xn + 0.15*np.random.standard_normal(len(xn))  # add random noise
yn = f(xn) + 0.25 * np.random.standard_normal(len(xn)) # y values with random noise
```

```python
reg = np.polyfit(xn, yn, 7)
ry = np.polyval(reg, xn)
```

```python
create_plot([xn,xn], [f(xn), ry], ['b', 'r.'], ['f(x)', 'd7_reg'], ['xn', 'f(x)'])
```
