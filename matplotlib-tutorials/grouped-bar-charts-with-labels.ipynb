## Grouped Bar Chart with Labels

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
data = tips.groupby(['day', 'sex'])['tip'].mean().unstack()
```

```python
data
```

```python
labels = data.index # Days
male = data["Male"]
female = data["Female"]
```

```python
x = np.arange(len(labels)) # location of x ticks
width = 0.4 # width of bars
```

```python
fig, ax = plt.subplots()
p1 = ax.bar(x - width/2, male, width, label='Male')
p2 = ax.bar(x + width/2, female, width, label='Female')

# Labels
ax.set_ylabel("Mean Tip")
ax.set_title("Mean Tip by Day and Gender")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

ax.bar_label(p1, padding=3, fmt="%.2f")
ax.bar_label(p2, padding=3, fmt="%.2f")
fig.tight_layout()

plt.show()
```
