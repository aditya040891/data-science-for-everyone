## Bar Chart Labels with Matplotlib (Horizontal bar charts)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(42)
```

```python
# Fake some data
company = ("META", "AMZN", "NFLX", "GOOG")
y_pos = np.arange(len(company))
performance = 4 + 10 * np.random.rand(len(company))
err = np.random.rand(len(company))

fig, ax = plt.subplots()
hbars = ax.barh(y_pos, performance, xerr=err, align="center")
ax.set_yticks(y_pos)
ax.set_yticklabels(company)
ax.invert_yaxis()  # labels top to bottom
ax.set_xlabel("Performance")
ax.set_title("FAANG Stock Peformance Q1")

ax.bar_label(hbars, labels=['±%.2f' % e for e in err], padding=10, color='r', fontsize=14)
ax.set_xlim(right=18)
plt.show()
```
