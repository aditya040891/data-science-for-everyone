## Error Bar Limit Selection

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
y = 2.5 * np.sin(x / 20 * np.pi)
yerr = np.linspace(0.05, 0.2, 10)
```

```python
fig = plt.figure(figsize=(10, 10))
plt.errorbar(x, y + 3, yerr = yerr, label = "both limits (default)")
plt.errorbar(x, y + 2, yerr = yerr, label = "uplims=True", uplims=True)
plt.errorbar(x, y + 1, yerr = yerr, label = "uplims=True, lolims=True", uplims=True, lolims=True)

upperlimits = [True, False] * 5
lowerlimits = [False, True] * 5
plt.errorbar(x, y, yerr = yerr, uplims=upperlimits, lolims=lowerlimits, label="subsets of uplims and lolims")
plt.legend(loc="lower right")
plt.show()
```
