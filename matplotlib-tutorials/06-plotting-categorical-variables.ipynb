## Plotting Categorical Variables Part 2

```python
import matplotlib.pyplot as plt
```

```python
boy = ["bored", "happy", "bored", "bored", "happy", "bored"]
girl = ["happy", "happy", "happy", "happy", "bored", "happy"]
```

```python
activities = ["breakfast", "reading", "game", "snack", "nap", "bathtime"]
```

```python
fig, ax = plt.subplots()
ax.plot(activities, boy, label="boy")
ax.plot(activities, girl, label="girl")
ax.legend()
plt.show()
```
