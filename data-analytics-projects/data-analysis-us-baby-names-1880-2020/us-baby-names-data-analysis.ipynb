```python
years = range(1880, 2021)
print(list(years))
```

```python
pieces = [] # list of dataframes
columns = ['name', 'gender', 'births']
for year in years:
    path = f'names/yob{year}.txt'
    frame = pd.read_csv(path, names=columns)
    frame['year'] = year
    pieces.append(frame)
```

```python
len(pieces)
```

```python
pieces
```

```python
names = pd.concat(pieces, ignore_index=True)
```

```python
names.head()
```

```python
names.tail()
```

```python
total_births = names.pivot_table('births', index='year', columns='gender', aggfunc=sum)
total_births.tail()
```

```python
total_births.plot(title='Total births by gender and year');
```

```python
names
```

```python
def add_prop(group):
    group['prop'] = group['births'] / group['births'].sum()
    return group
```

```python
names = names.groupby(['year', 'gender']).apply(add_prop).reset_index(drop=True)
```

```python
names.head()
```

```python
names.groupby(['year', 'gender'])['prop'].sum()
```

```python
def get_top1000(group):
    return group.sort_values(by='births', ascending=False)[:1000]
```

```python
grouped = names.groupby(['year', 'gender'])
```

```python
top1000 = grouped.apply(get_top1000)
```

```python
top1000 = top1000.reset_index(drop=True)
top1000
```

```python
dfm = male[male['year'] == 2020]
dfm.head()
```

```python
prop_cumsum = dfm.sort_values(by='prop', ascending=False)['prop'].cumsum()
prop_cumsum[:10]
```

```python
prop_cumsum.values.searchsorted(0.5) 
```

```python
def get_quantile_count(group, q=0.5):
    group = group.sort_values(by='prop', ascending=False)
    return group['prop'].cumsum().values.searchsorted(q) + 1
```

```python
diversity = top1000.groupby(['year', 'gender']).apply(get_quantile_count)
```

```python
diversity = diversity.unstack('gender')
diversity
```

```python
diversity.plot(title='Number of popular names in top 50%', figsize=(10, 10))
plt.show()
```
