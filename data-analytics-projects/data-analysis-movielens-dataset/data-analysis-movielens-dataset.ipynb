```python
import pandas as pd
```

```python
users = pd.read_csv('users.csv', sep=';')
ratings = pd.read_csv('ratings.csv', sep=';')
movies = pd.read_csv('movies.csv', sep=';', encoding='latin-1')
```

```python
for df in [users, ratings, movies]:
    print(df.head())
    print("-------------------------")
```

```python
data = pd.merge(pd.merge(ratings, users), movies)
```

```python
data.head()
```

```python
data = data.drop('Unnamed: 3', axis=1)
```

```python
data
```

```python
data.iloc[0]
```

```python
mean_ratings = data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
mean_ratings
```

```python
ratings_by_title = data.groupby('title').size()
ratings_by_title
```

```python
active_titles = ratings_by_title[ratings_by_title >= 250]
active_titles
```

```python
for title in active_titles.head(20).index:
    print(title)
```

```python
mean_ratings = mean_ratings.loc[active_titles.index]
mean_ratings
```

```python
top_female_ratings = mean_ratings.sort_values('F', ascending=False)
top_female_ratings
```

```python
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
```

```python
sorted_by_diff = mean_ratings.sort_values(by='diff')
sorted_by_diff.head(10)      # Women preferred movies
```

```python
sorted_by_diff[::-1].head(10) # Men preferred movies
```

```python
rating_std_by_title = data.groupby('title')['rating'].std()
```

```python
rating_std_by_title = rating_std_by_title.loc[active_titles.index]
```

```python
rating_std_by_title.sort_values(ascending=False).head(10)
```
